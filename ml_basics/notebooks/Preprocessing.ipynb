```python
import re
from nltk.corpus import stopwords # to remove the stopwords
from nltk.stem import PorterStemmer # to stem the words
import nltk # to remove the punctuation
```

```python
nltk.download('stopwords')
```

```python
print(stopwords.words('english'))
```

```python
text_data = pd.read_csv('/Users/alt/Library/CloudStorage/OneDrive-Personal/datas/train.csv')
text_data.sample(5)
```

```python
text_data = text_data.fillna('')
text_data['content'] = text_data['title'] + ' ' + text_data['author'] + ' ' + text_data['text']
```

```python
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def fast_stemming(text):
    words = re.sub('[^a-zA-Z]', ' ', text).lower().split()
    words = [stemmer.stem(word) for word in words if word not in stop_words]
    return ' '.join(words)

# Apply on the entire column
text_data['content'] = text_data['content'].map(fast_stemming) 
```

```python
vectorazer = TfidfVectorizer()
X = text_data['content'].values
Y = text_data['label'].values
```

```python
vectorazed_data = vectorazer.fit_transform(X)
```
